```python
from TVGL import *
```

```python
solver = TVGL(filename='Examples/Examples_data/perturbation.csv', lambd=0.01, beta=2, penalty_function="perturbed_node")
print(f"Blocks: {solver.blocks}")
print(f"Observations in a block: {solver.obs}")
print(f"Rho: {solver.rho}")
print(f"Lambda: {solver.lambd}")
print(f"Beta: {solver.beta}")
print(f"Penalty function: {solver.penalty_function}")
```

```python
solver.run_algorithm()
```

```python
solver.temporal_deviations()
deviations=solver.norm_deviations
```

```python
import matplotlib.pyplot as plt
plt.plot(deviations)
```
